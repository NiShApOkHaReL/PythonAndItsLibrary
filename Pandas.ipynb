## Merge() Function

```python
df1 = pd.DataFrame({'S.N':[1,2,3,4,5],'Name':['ABC','DEF','GHI','JKL','MNO']}) 
df1
```

```python
df2 = pd.DataFrame({'S.N':[1,2,3,4,5],'Marks':[30,45,78,54,22]}) 
df2
```

```python
pd.merge(df1,df2,on='S.N')
```

```python
pd.merge(df1,df2) #on  is by default intersecting column is taken
```

```python
df3 = pd.DataFrame({'S.N':[1,2,3,4,5], 'Name':['abc','def','daaa','grgr','uhy']})
df4 = pd.DataFrame({'S.N':[1,2,3,6,7],'Marks':[22,33,44,66,77]})
```

```python
pd.merge(df3,df4)
```

```python
pd.merge(df3,df4,how='left')
```

```python
pd.merge(df3,df4,how='right')
```

```python
pd.merge(df3,df4,how='outer')
```

## Append() function

```python
df1 = pd.DataFrame({'Roll No.':[1,2,3,4,5],'Maths':[22,34,56,22,45],'physics':[77,44,65,34,67]})
df2 = pd.DataFrame({'Roll No.':[6,7,8,9,10],'Maths':[20,17,98,90,55],'physics':[76,54,89,44,24]})
print(df1)
print(df2)
```

```python
df1.append(df2)
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2], ignore_index=True)
```

```python
pd.concat([df1,df2], ignore_index=True, sort = True)
```

```python
df1 = pd.DataFrame({'Roll No.':[1,2,3,4,5],'Nepali':[22,34,56,22,45],'physics':[77,44,65,34,67]})
df2 = pd.DataFrame({'Roll No.':[6,7,8,9,10],'Maths':[20,17,98,90,55],'physics':[76,54,89,44,24]})
pd.concat([df1,df2], ignore_index=True)
```
